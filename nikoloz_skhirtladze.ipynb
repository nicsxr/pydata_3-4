```python
import pandas as pd
```

```python
df = pd.read_csv("Earthquakes_database.csv")
df
```

```python
# 1
# გაერთიანება
df['Combined Magnitude'] = df['Magnitude'].astype(str) + ' ' + df['Magnitude Type'].astype(str)
df.drop(['Magnitude', 'Magnitude Type'], axis=1, inplace=True)
df

# გადანაწილება
df[['Month', 'Day', 'Year']] = df['Date'].str.split('/', expand=True)
df = df.drop(['Date'], axis=1)
df
```

```python
# 2
# კატეგორიებად დაყოფა სიმძლავრის მიხედვით
df = pd.read_csv("Earthquakes_database.csv")
bins = [0, 3, 6, 7, float('inf')]
labels = ['Low', 'Moderate', 'Strong', 'Very Strong']

df['Magnitude Class'] = pd.cut(df['Magnitude'], bins=bins, labels=labels, right=False)
df
```

```python
# 3
# pivot
pivot_df = pd.pivot_table(df, values='Magnitude', index='Magnitude Class', aggfunc='count', fill_value=0)
pivot_df

# melt
melted_df = pd.melt(df, id_vars=['Date', 'Time', 'Latitude', 'Longitude', 'Type', 'Depth', 'Magnitude Class'],
                    var_name='Attribute', value_name='Value')
melted_df
```

```python
# 4
import matplotlib.pyplot as plt


df = pd.read_csv("Earthquakes_database.csv")

df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Magnitude'], marker='o', linestyle='-')
plt.title('Magnitude over time')
plt.xlabel('Date')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()


magnitude_counts = df['Magnitude'].value_counts()
plt.bar(magnitude_counts.index, magnitude_counts.values, color='skyblue')

plt.xlabel('Magnitude')
plt.ylabel('Count')
plt.title('Earthquake Magnitudes')
```
